```python
# Importing the dataset into your code
from ucimlrepo import fetch_ucirepo

# fetch dataset
adult = fetch_ucirepo(id=2)

# data (as pandas dataframes)
X = adult.data.features
y = adult.data.targets

# metadata
print(adult.metadata)

# variable information
print(adult.variables)
```

```python
# Display the first few rows of the dataset
print(X.head())
```

```python
# Summary statistics for numerical features
print(X.describe())
```

```python
# Check for missing values
print(X.isnull().sum())
```

```python
# Assuming 'X' is your DataFrame
numerical_features = X.select_dtypes(include=['int64', 'float64'])

# Count missing values in each numerical column
missing_values_numerical = numerical_features.isnull().sum()

# Display the missing values count for numerical features
print(missing_values_numerical)
```

```python
# Assuming 'X' is your DataFrame
categorical_features = X.select_dtypes(include=['object'])

# Count missing values in each categorical column
missing_values_categorical = categorical_features.isnull().sum()

# Display the missing values count for categorical features
print(missing_values_categorical)
```

```python
# Check data types
print(X.dtypes)
```

```python
# Impute missing values with the median (for numerical columns) or mode (for categorical columns)
from sklearn.impute import SimpleImputer
import numpy as np

# Imputer for numerical data
num_imputer = SimpleImputer(strategy='median')
num_cols = X.select_dtypes(include=['int64', 'float64']).columns
X[num_cols] = num_imputer.fit_transform(X[num_cols])

# Imputer for categorical data
cat_imputer = SimpleImputer(strategy='most_frequent')
cat_cols = X.select_dtypes(include=['object']).columns
X[cat_cols] = cat_imputer.fit_transform(X[cat_cols])
```

```python
# Recheck missing values in each numerical column after handling
missing_values_numerical_after = numerical_features.isnull().sum()

# Display the missing values count for numerical features after handling
print("Missing Values in Numerical Columns after Handling:\n", missing_values_numerical_after)

# Recheck missing values in each categorical column after handling
missing_values_categorical_after = categorical_features.isnull().sum()

# Display the missing values count for categorical features after handling
print("Missing Values in Categorical Columns after Handling:\n", missing_values_categorical_after)
```

```python
# Extracting column names for numerical features
numerical_feature_names = numerical_features.columns
print(numerical_feature_names)
```

```python
# Calculate IQR for each numerical column
Q1 = X[numerical_feature_names].quantile(0.25)
Q3 = X[numerical_feature_names].quantile(0.75)
IQR = Q3 - Q1

# Determine outliers using IQR
outliers = ((X[numerical_feature_names] < (Q1 - 1.5 * IQR)) | (X[numerical_feature_names] > (Q3 + 1.5 * IQR)))
```

```python
# Filter the DataFrame to display only rows with outliers
outlier_rows = X[outliers.any(axis=1)]
print(outlier_rows)
```

```python
# Display outliers for each numerical column individually
for col in numerical_feature_names:
    if outliers[col].any():
        print(f"Outliers in {col}:")
        print(X.loc[outliers[col], col])
        print("\n")
```

```python
# Remove rows with outliers
X_cleaned = X[~(outliers.any(axis=1))]
```

```python
# Cap and floor outliers
X_cleaned = X.copy()  # Create a copy of the original DataFrame to preserve the original data

for col in numerical_feature_names:
    lower_bound = Q1[col] - 1.5 * IQR[col]
    upper_bound = Q3[col] + 1.5 * IQR[col]
    X_cleaned[col] = X_cleaned[col].clip(lower=lower_bound, upper=upper_bound)
```

```python
# Apply log transformation
import numpy as np

X_cleaned = X.copy()  # Create a copy of the original DataFrame for transformation

for col in numerical_feature_names:
    # Apply log transformation using np.log1p which is log(x + 1)
    # This helps in avoiding issues with zero or negative values in the data
    X_cleaned[col] = np.log1p(X_cleaned[col])
```

```python
# Verify by checking the statistical summary or plotting
print(X_cleaned.describe())
# or use plots like boxplots, histograms, etc.
```

```python
# Assuming 'X_cleaned' is your DataFrame after outlier handling
# and 'numerical_feature_names' are your numerical column names

from itertools import combinations

# Generate scatter plots for each pair of numerical features
for (col1, col2) in combinations(numerical_feature_names, 2):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=X_cleaned[col1], y=X_cleaned[col2])
    plt.title(f'Scatter Plot of {col1} vs {col2} After Outlier Handling')
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.show()
```

```python
# Identify categorical columns
# Assuming 'X' is your DataFrame

# Select categorical columns
categorical_features = X.select_dtypes(include=['object'])

# Get the names of the categorical columns
categorical_feature_names = categorical_features.columns

# Display the names of the categorical columns
print("Categorical Columns:")
print(categorical_feature_names)
```

```python
# One-Hot Encoding of categorical features
import pandas as pd
X_encoded = pd.get_dummies(X[categorical_feature_names])

# Combine the encoded categorical features with the cleaned numerical features
X_final = pd.concat([X_cleaned, X_encoded], axis=1)
```

```python
# Verify the structure of dataframe after encoding
print(X_final.head())
print(X_final.info())
```

```python
from sklearn.preprocessing import StandardScaler
# Checks for any non numeric data type
print(X_encoded.dtypes)

# Split the dataset
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Apply feature Scaling
scaler = StandardScaler()

# Fit the scaler on the training data and transform the training data
X_train_scaled = scaler.fit_transform(X_train)

# Transform the test data using the same scaler
X_test_scaled = scaler.transform(X_test)

# Convert scaled data back to dataframe
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)

# Check the statistical summary of the scaled training data
print(X_train_scaled.describe())
```

```python
# Check the distribution of the target variable
class_counts = y_train.value_counts()
print(class_counts)
```

```python
# Handling imbalanced data
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

# Apply SMOTE only on training data
X_train_balanced, y_train_balanced = smote.fit_resample(X_train_scaled, y_train)

# Assuming y_train_balanced is the target variable after applying SMOTE or another balancing technique

# Check the new class distribution directly
new_class_counts = y_train_balanced.value_counts()
print(new_class_counts)
```

```python
# Train the model with selected features
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# Create a model
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Assuming X_train_scaled and X_test_scaled are your data after feature scaling and potentially other preprocessing
# If you have a specific feature selection step, apply it here to create X_train_selected and X_test_selected
# For this example, I am using X_train_scaled and X_test_scaled directly
# X_train_selected = X_train_scaled  # This line is causing the issue
# X_test_selected = X_test_scaled   # This line is causing the issue

# Use the balanced data from SMOTE:
X_train_selected = X_train_balanced
X_test_selected = X_test_scaled  # X_test remains the same

# Train the model using the selected features from the training set
model.fit(X_train_selected, y_train_balanced)

# Make predictions
y_pred = model.predict(X_test_selected)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Train a model on the full feature set for comparison
model_full = RandomForestClassifier(n_estimators=100, random_state=42)
model_full.fit(X_train_balanced, y_train_balanced)

# Make predictions
y_pred_full = model_full.predict(X_test_scaled)

# Evaluate the model
accuracy_full = accuracy_score(y_test, y_pred_full)
print("Accuracy with all features:", accuracy_full)
print("\nClassification Report with all features:\n", classification_report(y_test, y_pred_full))
```

```python
# Creating Pipeline
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier

# Get the names of numerical and categorical features from X_train
# Update: Include 'uint8' for one-hot encoded features, and correctly select numerical columns
numerical_features = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X_train.select_dtypes(include=['uint8', 'object']).columns.tolist()  # Include 'uint8' for one-hot encoded columns

# If you have no 'object' or 'uint8' type columns after one-hot encoding (unlikely but possible):
if not categorical_features:
    # If there are no categorical features, use all columns as numerical
    categorical_features = []
    numerical_features = X_train.columns.tolist() # use all columns

# Numeric Transformer
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),  # Imputes missing values with the mean
    ('scaler', StandardScaler())  # Standardizes the data
])

# Categorical Pipeline
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),  # Handles missing values
    ('onehot', OneHotEncoder(handle_unknown='ignore'))  # One-hot encodes categorical variables
])

# Column Transformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Create a full pipeline with classifier
clf = Pipeline(steps=[('preprocessor', preprocessor),
                      ('classifier', RandomForestClassifier())])

# Train the model
clf.fit(X_train, y_train) # Now this should work because the columns are correctly defined.
```
